```python
import torch
import matplotlib.pyplot as plt
import seaborn as sb

sb.set()
```

```python
# (E(c) / 3500) * 0.45
# diff_ (c / 3500) * 0.45
```

```python
def get_weight(c):
    return ((c / 3500) * 0.45)
```

```python
c = torch.linspace(1000,1500,100)
w = get_weight(c)
```

```python
fig, ax = plt.subplots()

_ = ax.plot(c, w)
```
