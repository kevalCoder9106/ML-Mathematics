```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
x = np.array(data.iloc[:,0:4])
targets = data['species'].unique()

targets
```

```python
pca = PCA(n_components=2)
```

```python
X = pca.fit_transform(x)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=np.array(data['species'].replace(targets,[0,1,2])))
plt.show()
```
