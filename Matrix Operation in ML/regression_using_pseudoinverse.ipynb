```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('sat_gpa.csv')
```

```python
x = np.array([np.zeros(84) + 1,data['SAT']])
y = np.array([data['GPA']])

x.shape, y.shape
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[1], y.reshape(-1,1))
plt.show()
```

```python
x_ = np.linalg.pinv(x)
w = np.dot(x_.T, y.T)

b = w[0,0]
w = w[1,0]
```

```python
fig, ax = plt.subplots()

ax.scatter(x[1], y)

x_min, x_max = ax.get_xlim()
y_min, y_max = w * x_min + b, w * x_max + b

ax.plot([x_min, x_max],[y_min,y_max])
```
