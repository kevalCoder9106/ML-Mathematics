```python
import numpy as np
import cv2
```

```python
vid = cv2.VideoCapture('vid.mp4')
c_frame = 0
t_frame = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
w = int(vid.get(3))
h = int(vid.get(4))

fps = 30
sec = 5

fourcc = cv2.VideoWriter_fourcc(*'XVID')
v_writer = cv2.VideoWriter('output.mp4',fourcc,float(fps),(w,h))


while True:
    print("Current frame: ", c_frame, "/", t_frame)
    s, f = vid.read()

    r = f[:,:,0]
    g = f[:,:,1]
    b = f[:,:,2]

    feature_arr = []
    bits = 32

    for i in [r,g,b]:
        U, sigma, VT = np.linalg.svd(i)
        features = np.dot(U[:,:bits],np.dot(np.diag(sigma[:bits]),VT[:bits,:]))
        feature_arr.append(features)

    rec_f = np.dstack((feature_arr[0],feature_arr[1],feature_arr[2]))

    cv2.imwrite('image.png',rec_f)

    v_writer.write(cv2.imread('image.png'))

    if c_frame >= t_frame / 2:
        break

    c_frame += 1

v_writer.release()
```
