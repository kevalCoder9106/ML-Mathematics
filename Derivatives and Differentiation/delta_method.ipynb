```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

sb.set()
```

```python
x = np.linspace(-10,10,1000)
```

```python
def y_fun(x):
    y = x ** 2 + 2 * x + 2
    return y
```

```python
y = y_fun(x)
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x,y)
```

```python
x_1 = 2
y_1 = y_fun(x_1)
```

```python
fig, ax = plt.subplots()
plt.scatter(x_1, y_1)
_ = ax.plot(x,y)
```

```python
x_2 = 5
y_2 = y_fun(x_2)
```

```python
fig, ax = plt.subplots()
plt.scatter(x_1, y_1, zorder=3)
plt.scatter(x_2,y_2, zorder=3)
_ = ax.plot(x,y)
```

```python
m = (y_2 - y_1) / (x_2 - x_1)
```

```python
b = 10 - m * 2
```

```python
tang = x * m + b
```

```python
fig, ax = plt.subplots()

plt.scatter(x_1, y_1, zorder=3)
plt.scatter(x_2,y_2, zorder=3)
ax.plot(x, tang, c='orange')
_ = ax.plot(x, y)
```
