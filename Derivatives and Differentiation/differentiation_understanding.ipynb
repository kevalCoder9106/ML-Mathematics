```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

sb.set()
```

```python
def f(x):
    x = np.array(x)
    return -5 * (3 * (x ** 2))

def delta_method(x1, delta, f):
    x2 = x1 + delta
    y1 = f(x1)
    y2 = f(x2)

    return (y2-y1) / (x2-x1)

def diff_method(x1):
    return -15 * (x1 ** 2)
```

```python
x = np.linspace(-10,10,50)
y = f(x)
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x,y)
```

```python
a = []
b = []

for i in x:
    a.append(delta_method(i, 0.00001, f))
    b.append(diff_method(i))
```

```python
fig, ax = plt.subplots()
ax.plot(a, y)
ax.plot(b, y)
_ = ax.plot(x,y)
```
