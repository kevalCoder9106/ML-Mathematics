```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 2*x + x

def diff(f, x, delta):
    return (f(x + delta) - f(x)) / delta

def bias(m, x, y):
    return y - m * x

def plot_eq(b, m, x):
    return m * x + b
```

```python
x = np.linspace(-10,10,50)
y = f(x)
```

```python
fig, ax = plt.subplots()
x_arr = []
slope = []
index = 0
delta = 0.01

for i in x:
    if index > 4:
        ax.scatter(i,f(i), zorder=3)
        ax.scatter(i + delta,f(i + delta),c='black')
        slope.append(diff(f,i,delta))
        x_arr.append(i)
        index = 0
    index += 1

_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()

tang = x_arr[0] * slope[0] + bias(slope[0],x_arr[0],f(x_arr[0]))
ax.plot(x_arr[0],tang)

_ = ax.plot(x, y)
```
