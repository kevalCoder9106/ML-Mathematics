```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return (x ** 2) - (y ** 2)

def diff_wrt_x(x):
    return 2 * x

def diff_wrt_y(y):
    return -2 * y

def plot_tangent_x(xs, x, y, f, diff, col):
    z = f(x, y)
    plt.scatter(x, z, c=col, zorder=3)

    m = diff(x)
    b = z - m * x
    tangent = m * xs + b

    plt.plot(xs, tangent, c=col, linestyle='dashed')

def plot_tangent_y(ys, x, y, f, diff, col):
    z = f(x, y)
    plt.scatter(y, z, c=col, zorder=3)

    m = diff(y)
    b = z - m * y
    tangent = m * ys + b

    plt.plot(ys, tangent, c=col, linestyle='dashed')
```

```python
xs = np.linspace(-2, 2, 10)
x_sample = [-2, -1, 0, 1, 2]
col = ['red', 'blue', 'green', 'yellow', 'pink', 'purple']
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(2,-2)
plt.ylim(-1,5)

for x,c  in zip(x_sample, col):
    plot_tangent_x(xs, x, 0, f, diff_wrt_x, c)

_ = ax.plot(xs, f(xs, 0))
```

```python
ys = np.linspace(-2, 2, 10)
zs = f(0, ys)
y_sample = [-2, 1, 0, 1, 2]
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-4, 4)
plt.ylim(-5, 1)
for y, c in zip(y_sample, col):
    plot_tangent_y(ys, 0, y, f, diff_wrt_y, c)

_ = ax.plot(ys, zs)
```
